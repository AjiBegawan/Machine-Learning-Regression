```python
def set_fillna(df):
    null_feature = ['awake', 'caffeine', 'alcohol', 'exercise_freq']
    df[null_feature] = df[null_feature].fillna(df[null_feature].mean())
    return df

df = set_fillna(df)
```

```python
def set_types(df):
    df['id'] = df['id'].astype('int16')
    df['age'] = df['age'].astype('int8')
    df['gender'] = df['gender'].astype('category')
    df['bed_time'] = pd.to_datetime(df['bed_time'])
    df['wakeup_time'] = pd.to_datetime(df['wakeup_time'])
    df['sleep_dur'] = df['sleep_dur'].astype('float16')
    df['sleep_eff'] = df['sleep_eff'].astype('float16')
    df['rem_per'] = df['rem_per'].astype('int8')
    df['deep_per'] = df['deep_per'].astype('int8')
    df['light_per'] = df['light_per'].astype('int8')
    df['awake'] = df['awake'].astype('int8')
    df['caffeine'] = df['caffeine'].astype('int8')
    df['alcohol'] = df['alcohol'].astype('int8')
    df['smoke_stat']         = df['smoke_stat'].map({'Yes': 1, 'No':0}).astype('int8')
    df['exercise_freq'] = df['exercise_freq'].astype('int8')
    
    return df

df = set_types(df)
print(df.info())
df.describe()
```

```python
X = df[feature]
Y = df[target]

scaler = StandardScaler()
X = scaler.fit_transform(X)
# X = scaler.fit_transform(pca_data)

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 42)

print(f'Total # of sample in whole dataset: {len(X)}')
print(f'Total # of sample in train dataset: {len(X_train)}')
print(f'Total # of sample in test dataset: {len(X_test)}')
```

```python
def prepare_model(algorithm):
    model = algorithm
    model.fit(X_train, y_train.values.ravel())
    return model
```

```python
%%time
algorithms = [
#                 ('Linear Regression', LinearRegression()), 
#               ('SVR', SVR(kernel='rbf')), 
#               ('KNN', KNeighborsRegressor(n_neighbors=5)), 
#               ('Decision Tree', DecisionTreeRegressor(max_depth=14)),
#               ('GradientBoosting', GradientBoostingRegressor(n_estimators=11758, learning_rate=0.1, random_state=42)),
              ('MLP',MLPRegressor(hidden_layer_sizes=(1000000,), max_iter=100, random_state=42))
             ]

trained_models = []
model_and_score = {}
result = []

for index, tup in enumerate(algorithms):
    model  = prepare_model(tup[1])
    y_pred = model.predict(X_test)
    
    # Calculate the Accuracy, mean squared error (MSE), and R-squared score
    acc = str(round(model.score(X_train,y_train), 4)*100)+"%"
    mse = mean_squared_error(y_test, y_pred)
    r2  = r2_score(y_test, y_pred)
    
    model_and_score[tup[0]] = acc
    trained_models.append((tup[0],model))
    result.append((tup[0], acc, mse, r2))

results_df = pd.DataFrame(result, columns=('Model', 'Accuracy', 'MSE', 'R Square'))
results_df
```

```python
# Preparation
X = df[feature]
Y = df[target]

scaler = StandardScaler()
X = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)

print(f'Total # of sample in whole dataset: {len(X)}')
print(f'Total # of sample in train dataset: {len(X_train)}')
print(f'Total # of sample in test dataset: {len(X_test)}')

X_train = X_train.reshape((X_train.shape[0],X_train.shape[1],1))
X_test = X_test.reshape((X_test.shape[0],X_test.shape[1],1))


n_timesteps, n_features, n_outputs = X_train.shape[1], X_train.shape[2], y_train.shape[1]
print(n_timesteps)
print(n_features)
print(n_outputs)
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Dense, Flatten

# Define the model architecture
model = Sequential()
model.add(Conv1D(filters=32, kernel_size=2, activation='relu', input_shape=(n_timesteps, n_features)))
model.add(MaxPooling1D(pool_size=2))
model.add(Conv1D(filters=64, kernel_size=2, activation='relu'))
model.add(Flatten())
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='relu'))

model.summary()

# Compile the model
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])

# # Train the model
model.fit(X_train, y_train, epochs=10, batch_size=16, validation_data=(X_test, y_test), verbose = 2)

# Evaluate the model
mse = model.evaluate(X_test, y_test)
print("Mean Squared Error:", mse)

# # Make predictions
# y_pred = model.predict(X_test)
```
